```python
def growth_perc(x,y):
    growth = (y - x)/x
    growth = growth*100
    growth = growth.round(2)
    return growth

def total_growth(a,b):
    tot_growth = (b-a)/a
    tot_growth = tot_growth*100
    tot_growth = tot_growth.round(2)
    return tot_growth
```

```python
perc_df['2006 Growth %'] = growth_perc(perc_df['2005 GDP'],perc_df['2006 GDP'])
perc_df['2007 Growth %'] = growth_perc(perc_df['2006 GDP'],perc_df['2007 GDP'])
perc_df['2008 Growth %'] = growth_perc(perc_df['2007 GDP'],perc_df['2008 GDP'])
perc_df['2009 Growth %'] = growth_perc(perc_df['2008 GDP'],perc_df['2009 GDP'])
perc_df['2010 Growth %'] = growth_perc(perc_df['2009 GDP'],perc_df['2010 GDP'])
perc_df['2011 Growth %'] = growth_perc(perc_df['2010 GDP'],perc_df['2011 GDP'])
perc_df['2012 Growth %'] = growth_perc(perc_df['2011 GDP'],perc_df['2012 GDP'])
perc_df['2013 Growth %'] = growth_perc(perc_df['2012 GDP'],perc_df['2013 GDP'])
perc_df['2014 Growth %'] = growth_perc(perc_df['2013 GDP'],perc_df['2014 GDP'])
perc_df['2015 Growth %'] = growth_perc(perc_df['2014 GDP'],perc_df['2015 GDP'])
perc_df['2016 Growth %'] = growth_perc(perc_df['2015 GDP'],perc_df['2016 GDP'])
perc_df['2017 Growth %'] = growth_perc(perc_df['2016 GDP'],perc_df['2017 GDP'])
perc_df['2018 Growth %'] = growth_perc(perc_df['2017 GDP'],perc_df['2018 GDP'])
perc_df['2010-2018 Growth %'] = total_growth(perc_df['2010 GDP'],perc_df['2018 GDP'])
```
