```python
from matplotlib import pyplot as plt
from scipy import stats
import numpy as np
import pandas as pd
```

```python
pop_gdp = pd.read_csv("POPGDPMerged.csv")
pop_gdp.head()
```

```python
pop_gdp_df = pop_gdp.drop(columns=['Unnamed: 0','Unnamed: 0_x'])
pop_gdp_df.head()
```

```python
pop_gdp_change = pop_gdp_df.drop(columns=['2005 GDP','2006 GDP', '2007 GDP', '2008 GDP', '2009 GDP'])
pop_gdp_change.head()
```

```python
data_2013_total = pop_gdp_change.groupby("Industry")["2013 GDP"].sum()
```

```python
data_2014_total = pop_gdp_change.groupby("Industry")["2014 GDP"].sum()
```

```python
data_2015_total = pop_gdp_change.groupby("Industry")["2015 GDP"].sum()
```

```python
data_2016_total = pop_gdp_change.groupby("Industry")["2016 GDP"].sum()
```

```python
data_2017_total = pop_gdp_change.groupby("Industry")["2017 GDP"].sum()
```

```python
data_2018_total = pop_gdp_change.groupby("Industry")["2018 GDP"].sum()
```

```python
frames = [data_2013_total, data_2014_total, data_2015_total, data_2016_total, data_2017_total, data_2018_total]
```

```python
industry_sum_data = pd.concat([data_2013_total, data_2014_total, data_2015_total, data_2016_total, data_2017_total, data_2018_total], axis=1, ignore_index=False)
industry_sum_data
```

```python
industry_2019_df = pd.concat([data_2018_total], axis=1, ignore_index=False)
industry_2019_df
```

```python
industry_2019_df.sum()
```

```python
industry_sum = 81850845
industry_sum
```

```python
def growth_per(x,y):
        growth = x/y
        growth = growth*100
        growth = growth.round(2)
        return growth
```

```python
industry_2019_df['Contribution% To Total US Economy'] = growth_per(industry_2019_df['2018 GDP'],industry_sum)
industry_2019_df
```

```python
final_bar_chart = industry_2019_df.sort_values(by='2018 GDP', ascending=True)
final_bar_chart
```

```python
final_bar_chart = final_bar_chart.rename(index={'Other services (except government and government enterprises)': "Other Services", 'Administrative and support and waste management and remediation services': "Administrative, Waste Management", 'Real estate and rental and leasing':'Real estate, Rental, Leasing', 'Government and government enterprises':'Government', 'Finance and insurance':'Finance, Insurance', 'Professional, scientific, and technical services':'Professional, Scientific, Technical Services', 'Health care and social assistance':'Health Care, Social Assistance', 'Wholesale trade':'Wholesale Trade', 'Retail trade':'Retail Trade', 'Transportation and warehousing':'Transportation, Warehousing', 'Accommodation and food services':'Accommodation, Food Services','Management of companies and enterprises':'Management Of Companies and Enterprises', 'Mining, quarrying, and oil and gas extraction':'Mining, Quarrying, Oil/Gas Extraction', 'Educational services':'Educational Services', 'Arts, entertainment, and recreation':'Arts, Entertainment', 'Agriculture, forestry, fishing and hunting':'Agriculture, Forestry, Fishing, Hunting'})
```

```python
bar_chart = final_bar_chart.plot.barh(None, y='Contribution% To Total US Economy')
bar_chart.set_xlabel("Percentage Of 2018 GDP")
bar_chart.set_title('2018 Industry Size USA GDP')
```
